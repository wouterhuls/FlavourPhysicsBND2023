```python
import numpy as np
import matplotlib.pyplot as plt

# the distance between the slits and the screen
L = 10
# the distance between the slits
d = 1
# the wavelength
lamda = 0.5*d

# compute the amplitude as function of the coordinate along the screen
def amplitude(x):
  # distance
  Ltot = np.sqrt(L**2 + x**2)
  #phase
  phase = 2*np.pi*Ltot/lamda
  # total amplitude, but defined such that max amplitude is for x=0
  Aabs = L/Ltot
  Aphase = np.exp(1j*phase)
  return Aabs*Aphase

xmin = -20.0
xmax = +20.0

x = np.linspace(start=xmin,stop=xmax,num=1000)
A1 = amplitude(x+d/2)
A2 = amplitude(x-d/2)
Atot = A1 + A2
y = Atot*np.conj(Atot)
y1 = A1*np.conj(A1)
y2 = A2*np.conj(A2)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)







```

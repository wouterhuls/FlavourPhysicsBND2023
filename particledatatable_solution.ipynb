```python
# load the pdt
import pypdt
tbl = pypdt.ParticleDataTable()

# draw mass versus lifetime and mass versus width
import matplotlib.pyplot as plt

mass     = [ p.mass for p in tbl]  # GeV
width    = [ p.width for p in tbl] # GeV
fig, axes = plt.subplots(1,2)
fig.set_size_inches(10,4)
for ax in axes:
  ax.set_xlim([0.005, 300])
  ax.set_yscale("log")
  ax.set_xscale("log")
  ax.set_xlabel("mass [GeV]")

ax1 = axes[0]
ax2 = axes[1]

ax1.scatter( mass, width)
ax1.set_ylabel("width [GeV]")

#ax2.set_ylim([1e-22,1])

ax2.scatter( [p.mass for p in tbl if p.lifetime], [p.lifetime*1e-12 for p in tbl if p.lifetime])
ax2.set_ylabel("lifetime [s]")

# where are the weak decays? strong decays? electromagnetic decays?

```
